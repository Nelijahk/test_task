first, it is better to normalise all columns but target(to not do back scaling after predictions). therefore, all 53 columns will be in one range -> [0, 1].  

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
# scaling only for X
X_scaled = scaler.fit_transform(X)
```

the hidden_test.csv file is given without ground truth, therefore to evaluate the model quality the train data is divided on train and test subsets.

```python
from sklearn.model_selection import train_test_split

# 80% for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=666)
```

as the origin of data is unknown as well as their meaning it is not possible to select columns that influence the most on the result. so, it is started from simple LinearRegression model to understand if is needed to use smth more complex.

```python
from sklearn import linear_model
from sklearn.metrics import root_mean_squared_error

linear = linear_model.LinearRegression().fit(X_train, y_train)
y_pred = linear.predict(X_test)
root_mean_squared_error(y_test, y_pred)
```

not good results as it is already known that the target range is [0, 100] the rmse 28 means that on average prediction deviates by +-28. it is quite big for that range

as it is said in the instruction file it is needed to provide training file, so, it is probably expected model to be trained. therefore, next code is simple regression model with fully connected layers and relu activation func (to add non-linearity in model, otherwise it is just linear transformation regardless of the number of layers)

```python
import torch
from torch import nn, optim

class RegressionModel(nn.Module):
    def __init__(self, input_size):
        super(RegressionModel, self).__init__()

        self.layers = nn.Sequential(
            nn.Linear(input_size, 256),
            nn.ReLU(),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 1),
            # nn.ReLU(),
            # nn.Linear(16, 1)
        )

    def forward(self, x):
        return self.layers(x)
```

to start training it is needed to transform data into tensors and create dataloaders

```python
# as y is still in pandas type Serias it is needed to be transform into numpy array
y_train_np = y_train.values.reshape(-1, 1)
y_test_np = y_test.values.reshape(-1, 1)
type(y_train_np)
```

```python
from torch.utils.data import Dataset, DataLoader

class RegressionDataset(Dataset):
    def __init__(self, X_data, y_data):
        self.X_data = torch.tensor(X_data, dtype=torch.float32)
        self.y_data = torch.tensor(y_data, dtype=torch.float32)

    def __len__(self):
        return len(self.X_data)

    def __getitem__(self, index):
        return self.X_data[index], self.y_data[index]
```

```python
train_dataset = RegressionDataset(X_train, y_train_np)
test_dataset = RegressionDataset(X_test, y_test_np)
train_dataset
```

```python
batch_size = 64

train_loader = DataLoader(
    dataset=train_dataset,
    batch_size=batch_size,
    # after each epoch data is shuffled
    shuffle=True
)

test_loader = DataLoader(
    dataset=test_dataset,
    batch_size=batch_size,
    shuffle=False
)
```

```python
# X_train.shape[1] is number of columms == input size of model
reg_model = RegressionModel(X_train.shape[1])

# loss func for regression mean squared error
criterion = nn.MSELoss()

# optimiser with learning rate
optimiser = optim.Adam(reg_model.parameters(), lr=0.0001)
```

```python
epochs = 50
for epoch in range(epochs):
    reg_model.train()
    for X_batch, y_batch in train_loader:
        y_pred = reg_model(X_batch)
        # now it is rmse loss
        loss = torch.sqrt(criterion(y_pred, y_batch))
        optimiser.zero_grad()
        loss.backward()
        optimiser.step()

    if (epoch + 1) % 5 == 0:
        print(f'Epoch [{epoch + 1}/{epochs}], Loss RMSE: {loss.item():.6f}')
```

```python
import numpy as np

reg_model.eval()
all_y_true = []
all_y_pred = []
with torch.no_grad():
    for X_batch, y_batch in test_loader:
        y_pred = reg_model(X_batch)
        all_y_true.append(y_batch.numpy())
        all_y_pred.append(y_pred.numpy())

y_test_final = np.concatenate(all_y_true)
y_pred_final = np.concatenate(all_y_pred)

rmse = root_mean_squared_error(y_test_final, y_pred_final)
print(f"Test RMSE : {rmse:.4f}")
```
